```python
import numpy as np
import cv2
import os
import moviepy.editor as moviepy
import matplotlib.pyplot as plt
import math
```

```python
album = cv2.imread('tarkan.jpg')
album = cv2.cvtColor(album, cv2.COLOR_BGR2RGB)
print(np.shape(album))
```

```python
angle = math.radians(60)
x0 = album.shape[0]/2-1
y0 = album.shape[1]/2-1
x0 = int(x0)
y0 = int(y0)
print(album.shape)
M = np.array([[math.cos(angle), -math.sin(angle), x0*(1-math.cos(angle))+ y0*math.sin(angle)],
              [math.sin(angle), math.cos(angle), y0*(1-math.cos(angle))- x0*math.sin(angle)]])
```

```python
result = cv2.warpAffine(album, M,(album.shape[1],album.shape[0]))

plt.imshow(result)
```
